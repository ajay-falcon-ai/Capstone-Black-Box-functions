```python
import numpy as np
from sklearn.inspection import permutation_importance
from sklearn.model_selection import train_test_split
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF
```

```python
# Load initial input data
inputs = np.load('initial_inputs.npy', allow_pickle=True)
# Load initial output data
outputs = np.load('initial_outputs.npy', allow_pickle=True)
```

```python
X = inputs
y = outputs
```

```python
kernel = RBF(length_scale=0.1, length_scale_bounds='fixed')
gp = GaussianProcessRegressor(kernel=kernel, alpha=1e-10)
gp.fit(X, y)

for i in range(1,21):
    # Select next query point using Expected Improvement
    # Permutation importance
    result = permutation_importance(gp, X, y, n_repeats=i, random_state=42)
    print(f"Iteration with {i} repeats:")
    # Display results
    for i in range(X.shape[1]):
        print(f"Feature {i}: Mean importance = {result.importances_mean[i]:.4f}")
    print("\n")
```
