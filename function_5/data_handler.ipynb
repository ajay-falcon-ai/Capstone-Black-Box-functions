```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 20)  # or any number you prefer
pd.set_option('display.max_columns', None)  # show all columns without wrapping

# Load initial input data
inputs = np.load('initial_inputs.npy', allow_pickle=True)
# Load initial output data
outputs = np.load('initial_outputs.npy', allow_pickle=True)
```

```python
# Track sources
source_labels = ['file'] * len(inputs)
```

```python
#Append week-1 data.
#Function 5:	[0.183674, 0.857143, 0.816327, 0.877551]
inputs = np.append(inputs, [[0.183674, 0.857143, 0.816327, 0.877551]], axis=0)
#Function 5:	872.3074873002278
outputs = np.append(outputs, [872.3074873002278], axis=0)
```

```python
source_labels.append('week-1')
```

```python
#Append week-2 data.
#Function 5:	[0.272727, 0.848484, 0.848485, 0.818182]
inputs = np.append(inputs, [[0.272727, 0.848484, 0.848485, 0.818182]], axis=0)
#Function 5:	743.1182481623213
outputs = np.append(outputs, [743.1182481623213], axis=0)
```

```python
source_labels.append('week-2')
```

```python
# Create DataFrame
df = pd.DataFrame(inputs, columns=['var-1', 'var-2', 'var-3', 'var-4'])
df['yield'] = outputs
df['source'] = source_labels

# Sort by transformed-side effect (larger is better)
df_sorted = df.sort_values(by='yield', ascending=False)

# Preview
print(df_sorted)
```
