```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 20)  # or any number you prefer
pd.set_option('display.max_columns', None)  # show all columns without wrapping

# Load initial input data
inputs = np.load('initial_inputs.npy', allow_pickle=True)
# Load initial output data
outputs = np.load('initial_outputs.npy', allow_pickle=True)
```

```python
# Track sources
source_labels = ['file'] * len(inputs)
```

```python
#Append week-1 data.
#Function 7:	[0.000000, 0.500000, 0.214286, 0.214286, 0.428572, 0.714286]
inputs = np.append(inputs, [[0.000000, 0.500000, 0.214286, 0.214286, 0.428572, 0.714286]], axis=0)
#Function 7:	1.181422704250112
outputs = np.append(outputs, [1.181422704250112], axis=0)
```

```python
source_labels.append('week-1')
```

```python
#Append week-2 data.
#Function 7:	[0.060606, 0.484848, 0.242424, 0.181818, 0.484848, 0.727273]
inputs = np.append(inputs, [[0.060606, 0.484848, 0.242424, 0.181818, 0.484848, 0.727273]], axis=0)
#Function 7:	0.9769171261966855
outputs = np.append(outputs, [0.9769171261966855], axis=0)
```

```python
source_labels.append('week-2')
```

```python
# Create DataFrame
df = pd.DataFrame(inputs, columns=['hype-para-1', 'hype-para-2', 'hype-para-3', 'hype-para-4', 'hype-para-5', 'hype-para-6'])
df['score'] = outputs
df['source'] = source_labels

# Sort by transformed-side effect (larger is better)
df_sorted = df.sort_values(by='score', ascending=False)

# Preview
print(df_sorted)
```
