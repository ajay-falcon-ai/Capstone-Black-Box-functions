```python
import numpy as np
import pandas as pd


# Load initial input data
inputs = np.load('initial_inputs.npy', allow_pickle=True)
# Load initial output data
outputs = np.load('initial_outputs.npy', allow_pickle=True)
```

```python
# Track sources
source_labels = ['file'] * len(inputs)
```

```python
#Append week-1 data.
#Function 3:	[0.464647, 0.666667, 0.171718]
inputs = np.append(inputs, [[0.464647, 0.666667, 0.171718]], axis=0)
#Function 3:	-0.11118571432117087
outputs = np.append(outputs, [-0.11118571432117087], axis=0)
```

```python
source_labels.append('week-1')
```

```python
#Append week-2 data.
#Function 3:	[0.303031, 0.242424, 0.474748]
inputs = np.append(inputs, [[0.303031, 0.242424, 0.474748]], axis=0)
#Function 3:	-0.030680318237259866
outputs = np.append(outputs, [-0.030680318237259866], axis=0)
```

```python
source_labels.append('week-2')
```

```python
# Create DataFrame
df = pd.DataFrame(inputs, columns=['compound-1', 'compound-2', 'compound-3'])
df['trans-side-effect'] = outputs
df['source'] = source_labels

# Sort by transformed-side effect (larger is better)
df_sorted = df.sort_values(by='trans-side-effect', ascending=False)

# Preview
print(df_sorted)
```
