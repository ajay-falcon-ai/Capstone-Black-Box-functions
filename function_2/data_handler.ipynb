```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 20)  # or any number you prefer
pd.set_option('display.max_columns', None)  # show all columns without wrapping

# Load initial input data
inputs = np.load('initial_inputs.npy', allow_pickle=True)
# Load initial output data
outputs = np.load('initial_outputs.npy', allow_pickle=True)
```

```python
# Track sources
source_labels = ['file'] * len(inputs)
```

```python
#Append week-1 data.
#Function 2:	[0.818182, 0.939394]
inputs = np.append(inputs, [[0.818182, 0.939394]], axis=0)
#Function 2:	0.051646432712366415
outputs = np.append(outputs, [0.051646432712366415], axis=0)
```

```python
source_labels.append('week-1')
```

```python
#Append week-2 data.
#Function 2:	[0.616162, 1.000000]
inputs = np.append(inputs, [[0.616162, 1.000000]], axis=0)
#Function 2:	0.18011924802681153
outputs = np.append(outputs, [0.18011924802681153], axis=0)
```

```python
source_labels.append('week-2')
```

```python
# Create DataFrame
df = pd.DataFrame(inputs, columns=['number-1', 'number-2'])
df['log-likelyhood-score'] = outputs
df['source'] = source_labels

# Sort by transformed-side effect (larger is better)
df_sorted = df.sort_values(by='log-likelyhood-score', ascending=False)

# Preview
print(df_sorted)
```
