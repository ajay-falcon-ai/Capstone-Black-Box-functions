```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 20)  # or any number you prefer
pd.set_option('display.max_columns', None)  # show all columns without wrapping

# Load initial input data
inputs = np.load('initial_inputs.npy', allow_pickle=True)
# Load initial output data
outputs = np.load('initial_outputs.npy', allow_pickle=True)
```

```python
# Track sources
source_labels = ['file'] * len(inputs)
```

```python
#Append week-1 data.
#Function 4:	[0.469388, 0.448980, 0.469388, 0.204082]
inputs = np.append(inputs, [[0.469388, 0.448980, 0.469388, 0.204082]], axis=0)
#Function 4:	-4.620232320523549
outputs = np.append(outputs, [-4.620232320523549], axis=0)
```

```python
source_labels.append('week-1')
```

```python
#Append week-2 data.
#Function 4:	[0.515152, 0.484848, 0.363636, 0.181819]
inputs = np.append(inputs, [[0.515152, 0.484848, 0.363636, 0.181819]], axis=0)
#Function 4:	-4.939960051262183
outputs = np.append(outputs, [-4.939960051262183], axis=0)
```

```python
source_labels.append('week-2')
```

```python
# Create DataFrame
df = pd.DataFrame(inputs, columns=['hype-para-1', 'hype-para-2', 'hype-para-3', 'hype-para-4'])
df['opt-place-score'] = outputs
df['source'] = source_labels

# Sort by transformed-side effect (larger is better)
df_sorted = df.sort_values(by='opt-place-score', ascending=False)

# Preview
print(df_sorted.head())
```
