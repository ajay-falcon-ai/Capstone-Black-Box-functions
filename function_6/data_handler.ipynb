```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 20)  # or any number you prefer
pd.set_option('display.max_columns', None)  # show all columns without wrapping

# Load initial input data
inputs = np.load('initial_inputs.npy', allow_pickle=True)
# Load initial output data
outputs = np.load('initial_outputs.npy', allow_pickle=True)
```

```python
# Track sources
source_labels = ['file'] * len(inputs)
```

```python
#Append week-1 data.
#Function 6:	[0.733334, 0.200000, 0.666667, 0.733334, 0.133334]
inputs = np.append(inputs, [[0.733334, 0.200000, 0.666667, 0.733334, 0.133334]], axis=0)
#Function 6:	-0.5377942602933444
outputs = np.append(outputs, [-0.5377942602933444], axis=0)
```

```python
source_labels.append('week-1')
```

```python
#Append week-2 data.
#Function 6:	[0.757576, 0.101010, 0.707071, 0.757576, 0.151515]
inputs = np.append(inputs, [[0.757576, 0.101010, 0.707071, 0.757576, 0.151515]], axis=0)
#Function 6:	-0.7102415173770004
outputs = np.append(outputs, [-0.7102415173770004], axis=0)
```

```python
source_labels.append('week-2')
```

```python
# Create DataFrame
df = pd.DataFrame(inputs, columns=['floor', 'sugar', 'eggs', 'butter', 'milk'])
df['flavour-score'] = outputs
df['source'] = source_labels

# Sort by transformed-side effect (larger is better)
df_sorted = df.sort_values(by='flavour-score', ascending=False)

# Preview
print(df_sorted)
```
