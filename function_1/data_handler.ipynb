```python
import numpy as np
import pandas as pd

pd.set_option('display.max_rows', 20)  # or any number you prefer
pd.set_option('display.max_columns', None)  # show all columns without wrapping

# Load initial input data
inputs = np.load('initial_inputs.npy', allow_pickle=True)
# Load initial output data
outputs = np.load('initial_outputs.npy', allow_pickle=True)
```

```python
# Track sources
source_labels = ['file'] * len(inputs)
```

```python
#Append week-1 data.
#Function 1: input	[0.848485, 0.777778]
inputs = np.append(inputs, [[0.848485, 0.777778]], axis=0)
#Function 1:	-1.2487829338434956e-49
outputs = np.append(outputs, [-1.2487829338434956e-49], axis=0)
```

```python
source_labels.append('week-1')
```

```python
#Append week-2 data.
#Function 1:	[0.404040, 0.959596]
inputs = np.append(inputs, [[0.404040, 0.959596]], axis=0)
#Function 1:	-9.343447791332904e-112
outputs = np.append(outputs, [-9.343447791332904e-112], axis=0)
```

```python
source_labels.append('week-2')
```

```python
# Create DataFrame
df = pd.DataFrame(inputs, columns=['cont-src-1', 'cont-src-2'])
df['radiation-reading'] = outputs
df['source'] = source_labels

# Sort by transformed-side effect (larger is better)
df_sorted = df.sort_values(by='radiation-reading', ascending=False)

# Preview
print(df_sorted)
```
